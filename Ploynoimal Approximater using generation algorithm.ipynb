```python
import numpy as np
import sys
from copy import deepcopy

exampleNo=1000
testNo=100

No_of_model_per_gen=1000
No_of_best_model_per_gen=100
# No_of_children_per_model=int(No_of_model_per_gen/No_of_best_model_per_gen)-1 # rember it must be a int
No_of_children_per_model=int(No_of_model_per_gen/No_of_best_model_per_gen)
no_of_genrations=500

size=[1,20,1]

inf=99999999999999999
```

```python
def ploynoimal(x):
  return  (x*x*x*x)-3*(x*x*x)
```

```python
labelSet=[]
featureSet=[]
for i in range(exampleNo):
  x=np.random.randint(1,10000)/1000
  labelSet.append(ploynoimal(x))
  featureSet.append(x)

test_labelSet=[]
test_featureSet=[]
for i in range(exampleNo):
  x=np.random.randint(1,10000)/1000
  test_labelSet.append(ploynoimal(x))
  test_featureSet.append(x)
```

function to weigths and baises

```python
def createWeigths(size):
  weigths=[]
  for layer in range(1,len(size)):
    Layer=[]
    for node in range(size[layer]):
      Node=[]
      for backNode in range(size[layer-1]):
        Node.append(np.random.randint(1,100)/10)
      Layer.append(Node)
    weigths.append(Layer)
  return weigths

def createBaises(size):
  baises=[]
  for layer in range(1,len(size)):
    Layer=[]
    for node in range(size[layer]):
      Layer.append(np.random.randint(1,100)/10)
    baises.append(Layer)
  return baises
```

```python
networks_weigths=[createWeigths(size) for i in range(No_of_model_per_gen)]
networks_baises=[createBaises(size) for i in range(No_of_model_per_gen)]
```

activation and network running function

```python
def ReLu(x):
  return abs(x)
def runNN(weigths,baises,input):
  netwrokRun=[[input]]
  for layer in range(len(weigths)):
    Layer=[]
    for frontNode in range(len(weigths[layer])):
      Layer.append( ReLu( np.dot( netwrokRun[layer],weigths[layer][frontNode])+baises[layer][frontNode] ) )
    netwrokRun.append(Layer)
  return int(netwrokRun[len(netwrokRun)-1][0])
```

cost function for a given network

```python
def cost(weigths,baises):
  cost=0
  for example in range(exampleNo):
    cost+=abs( runNN(weigths,baises,featureSet[example])-labelSet[example] )
  return cost
```

temp function

```python
def find_max_index(array):
  max_index = 0
  for i in range(1, len(array)):
    if array[i] > array[max_index]:
      max_index = i
  return max_index
```

modification functions

```python
def modifyWeigths(weigths,g):
  if(g>0):
    sign=[-1,1]
    for layer in range(len(weigths)):
      for frontNode in range(len(weigths[layer])):
        for node in range(len(weigths[layer][frontNode])):
          if(np.random.randint(1,11)<=5):
            weigths[layer][frontNode][node]+=sign[np.random.randint(0,2)]*np.random.randint(1,100)/100
          else:
            weigths[layer][frontNode][node]+=sign[np.random.randint(0,2)]*np.random.randint(1,301)/10
  return weigths
def modifyBaises(baises,g):
  if(g>0):
    sign=[-1,1]
    for layer in range(len(baises)):
      for frontNode in range(len(baises[layer])):
        if(np.random.randint(1,11)<=5):
          baises[layer][frontNode]+=sign[np.random.randint(0,2)]*np.random.randint(1,100)/100
        else:
          baises[layer][frontNode]+=sign[np.random.randint(0,2)]*np.random.randint(1,301)/10
  return baises
```
